1) Instruction( For Data cleaning)
   - Find all Null values in the dataset. If there is any null value in any column, then fill it with the mean of that column.

```python
df.drop_duplicates(inplace=True) #in df, checks for each row with another one on whole column basis.
df.reset_index(inplace=True)
# df.drop(columns=['index','level_0'],inplace=True) #In drop, inplace is required.
```
